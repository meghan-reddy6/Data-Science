```python
#Concatenate two tuples: (1, 2, 3) and (4, 5, 6). Print the resulting tuple.
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
resulting_tuple = tuple1 + tuple2
print("Concatenated tuple:", resulting_tuple)
```

```python
#Create a tuple with duplicate elements and count the occurrences of an element. Find the index of the first occurrence of an element in the tuple.

tuple_with_duplicates = (1, 2, 3, 2, 4, 2, 5)
element_to_count = 2
count = tuple_with_duplicates.count(element_to_count)
index = tuple_with_duplicates.index(element_to_count)
print(f"Element {element_to_count} occurs {count} times.")
print(f"First occurrence of element {element_to_count} is at index {index}.")
```

```python
#Create a tuple containing 3 tuples, each with 3 elements. Print the tuple of tuples.

tuple_of_tuples = ((1, 2, 3), (4, 5, 6), (7, 8, 9))
print("Tuple of tuples:", tuple_of_tuples)
```

```python
#Create a tuple with the first 5 positive integers. Convert it to a list, append the number 6, and convert it back to a tuple. Print the resulting tuple.
initial_tuple = (1, 2, 3, 4, 5)
temp_list = list(initial_tuple)
temp_list.append(6)
resulting_tuple = tuple(temp_list)
print("Resulting tuple after appending 6:", resulting_tuple)
```

```python
#Create a tuple with the characters of a string. Join the tuple elements into a single string. Print the string.

tup = ('h', 'e', 'l', 'l', 'o')
print("".join(tup))
```

```python
#Create a dictionary with tuple keys and integer values. Print the dictionary.
tuple_dict = {('a', 1): 10, ('b', 2): 20, ('c', 3): 30}
print("Dictionary with tuple keys:", tuple_dict)
```

```python
#Create a nested tuple and iterate over the elements, printing each element.
nested_tuple = ((1, 2), (3, 4), (5, 6))
for inner_tuple in nested_tuple:
    for element in inner_tuple:
        print(element)
```

```python
#Create a tuple with duplicate elements. Convert it to a set to remove duplicates and print the resulting set.

tuple_with_duplicates = (1, 2, 2, 3, 4, 4, 5)
print(tuple(set(tuple_with_duplicates)))
```

```python
#Write functions that take a tuple and return the minimum, maximum, and sum of the elements. Print the results for a sample tuple.
def tuple_min(tup):
    return min(tup)
def tuple_max(tup):
    return max(tup)
def tuple_sum(tup):
    return sum(tup)

sample_tuple = (3, 1, 4, 1, 5, 9, 2, 6, 5)
print("Minimum:", tuple_min(sample_tuple))
print("Maximum:", tuple_max(sample_tuple))
print("Sum:", tuple_sum(sample_tuple))
```
